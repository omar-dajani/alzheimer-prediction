```python
import os
import pandas as pd
import re
from pathlib import Path
```

```python
alpha_path = Path("../data/01_raw/ADNI/ADNIAlpha")
beta_path = Path("../data/01_raw/ADNI/ADNIBeta")
```

```python
def get_unique_ptids(base_path):
    """
    Scans a directory and returns a set of unique patient IDs
    """
    ptids = set()
    if not base_path.exists():
        print(f"Can't find {base_path}.")
        return ptids

    # Regex for ADNI format: 3 digits, S, 4 digits
    ptid_pattern = re.compile(r'\d{3}_S_\d{4}')

    print(f"Looking into {base_path}")
    for root, dirs, files in os.walk(base_path):
        # Check path for PTID pattern
        # We search the full path string
        match = ptid_pattern.search(str(root))
        if match:
            ptids.add(match.group(0))
            
    return ptids
```

```python
# Get sets of patients
alpha_patients = get_unique_ptids(alpha_path)
beta_patients = get_unique_ptids(beta_path)
```

```python
# Perform set operations
overlapping_patients = sorted(list(alpha_patients.intersection(beta_patients)))
alpha_only_patients = sorted(list(alpha_patients - beta_patients))
beta_only_patients = sorted(list(beta_patients - alpha_patients))
```

```python
# 3. Create a DataFrame for easy viewing/saving
# We create a dictionary to handle uneven list lengths
max_len = max(len(overlapping_patients), len(alpha_only_patients), len(beta_only_patients))

def pad_list(lst, length):
    return lst + [None] * (length - len(lst))

df_lists = pd.DataFrame({
    'Overlapping_PTIDs': pad_list(overlapping_patients, max_len),
    'Alpha_Only_PTIDs': pad_list(alpha_only_patients, max_len),
    'Beta_Only_PTIDs': pad_list(beta_only_patients, max_len)
})
```

```python

print("Patient ID List")
print(f"Overlapping patients ({len(overlapping_patients)})")
print("Patients present in BOTH ADNI3 Alpha and ADNI4 Beta:")
print(df_lists['Overlapping_PTIDs'].dropna().tolist())
if len(overlapping_patients) > 15:
    print(f"and {len(overlapping_patients) - 15} more")

print(f"Alpha Only Patients ({len(alpha_only_patients)})")
print(df_lists['Alpha_Only_PTIDs'].dropna().tolist())

print(f"Beta Only Patients ({len(beta_only_patients)})")
print(df_lists['Beta_Only_PTIDs'].dropna().tolist())

# Save to CSV for the next steps
# output_file = "adni_patient_groups.csv"
# df_lists.to_csv(output_file, index=False)
# print(f"\nSuccessfully saved detailed patient lists to: {output_file}")
```
